```python
import matplotlib.pyplot as plt
import numpy as np
import math
import cmath

FREQUENCY_RANGE = 1000     #change this to max freq in Hz

def V_o(frequency):
    '''
    DESCRIPTION: Function that computes the Output Voltage from a transfer function
    -------------
    INPUTS: FREQUENCY             INTEGAR
    UNITS: HERTZ
    -------------
    OUTPUTS: OUTPUT VOLTAGE       INTEGAR
    UNITS: VOLTS
    -------------
    
    '''
    s = complex(0, 2 * math.pi * frequency)
    Hf = (10 * s + 1) / (s * (0.01 * s + 2)) # replace this with your own function
    return Hf


def main (x_axis = FREQUENCY_RANGE):
    '''
    DESCRIPTION: MAIN BODE PLOT FUNCTION
    -------------
    INPUTS: FREQUENCY (X-AXIS)      INTEGER     
        
    UNITS: HERTZ
    -------------
    OUTPUTS: MAGNITUDE AND PHASE PLOTS
    UNITS: N/A
    -------------
    
    '''
    
    #CREATES A RANGE OF FREQUENCIES TO MAXIMUM FREQUENCY
    values = np.arange(1, x_axis+1).tolist()
    y_values = list(map(V_o, values))
    x_values = np.arange(1,x_axis+1).tolist()
    
    #CREATES A PLOT OF Vo VS. FREQUENCY
    ig, magnitude = plt.subplots()
    magnitude.plot(list(abs(x) for x in x_values), y_values)
    magnitude.set(
        xlabel="Frequency (rads/s)", 
        ylabel='Magnitude (dB)', 
        title= 'Magnitude'
    )
    magnitude.set_xscale('log')
    magnitude.grid(which ='both')
    
    #COMPUTES THE PHASE AND PLOTS THE PHASE VS. FREQUENCY
    ih, phase = plt.subplots()
    phase.plot(x_values, list(math.degrees(x) for x in list(map(cmath.phase, y_values))))
    phase.set(
        xlabel="Frequency (rad/s)", 
        ylabel='Phase (degrees)', 
        title= 'Phase'
    )
    phase.set_xscale('log')
    phase.grid(which ='both')

if __name__ == '__main__':
    main()
```

## Notes
* map is a function that enables you to push an iterable (ie list) to a function -avoiding 
ineffecient iteration
* abs() when applied to complex numbers, gives you the magnitude
* np.arrange was used instead of range() as it enables you to use a rational number for the the step size
* further efficencies could be created by using itertools library to create an iterator instead of a list, conserving memory until code is executed.
